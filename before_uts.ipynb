# Membuat fungsi Sinus

Fungsi sinus dapat dibuat menggunakan Library Numpy, sbb:

```python
import numpy as np
import matplotlib.pyplot as plt
import math
t = np.arange(0,100)*0.02
d = np.zeros(100)
d = np.sin(2 * math.pi * t)
plt.plot(t,d)
plt.show()
```
